Analisis Exploratorio de Datos con enfoque en la recaudacion de peliculas

En este analisis exploratorio de datos se examino en onjunto de datos relacionados con .fotogramas. con enfoque en el listado de peliculas con mayor recudacion de este año. Esto con el prposito de reisitar estos acontecimiento del septimo arte que comprender que clase de peliculas son la que tienen mayor impacto en la audiencia general.

```python
peliculas = ['Super Mario Bros: La película', 'Barbie', 'Fast and Furious X', 'La sirenita', 'Guardianes de la Galaxia vol. 3']
ventas = [27, 16, 12, 11, 11]
```

```python
fig, ax = plt.subplots()
ax.set_ylabel('Ventas (Millones)')
ax.set_title('Taquillas mundiales')
plt.bar(peliculas, ventas)
plt.savefig('barras_simple.png')
plt.show()
```

Como se puede apresiar en la table anteriormente mostrada, en el top 5 de las peliculas con mayor recaudacion a nivel mundial hasta ahora solo se encuentran peliculas de grandes empresas como lo son Universal, Disney, Mervel, etc. Con esto podemos asumir que estas grandes compañias tienen un gran impacto en este campo del entretenimiento. en la cime de la piramide se encuentra una de las peliculas mas esperadoas de los ultimos años por los fans, como lo es Supr Mario Bros.
